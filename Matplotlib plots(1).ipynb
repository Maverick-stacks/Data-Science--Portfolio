```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Linear Plot
x = np.arange(1,10)
y = x ** 2
plt.plot(x,y)
plt.title("Linear Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

```python
#BarChart
x = ['Apples', 'Bananas', 'Cherries']
y = [10, 20, 15]
plt.bar(x, y, color= 'orange')
plt.title("Fruit Chart")
plt.xlabel("Fruits")
plt.ylabel("Quantity")
plt.grid(True)
plt.show()
```
