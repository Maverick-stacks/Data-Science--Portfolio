```python
#Create a line and scatter combo plot
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x,y, label='Line')
plt.scatter(x,y, color= 'red', s=10, label='Points')
plt.title("Sine Waves")
plt.xlabel("X Values")
plt.ylabel("Sin(x)")
plt.grid(True)
plt.legend()
plt.show()
```
