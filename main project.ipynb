```python
#USING SEABORN, MATPLOTLIB, PANDAS AND NUMPY TO CLEAN AND ILLUSTRATE DATA

#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the sample dataset
df = sns.load_dataset("tips")
```

```python
#Previewing the dataset
df.head(10)
```

```python
df.info()
```

```python
#Basic Summary
print("\nStatistics:\n",df.describe())
```

```python
df.shape
```

```python
#Data Cleaning#

#checking for duplicates
df.duplicated().sum()

```

```python
#checking for null/ missing values
df.isnull()
```

```python
#Using numpy and pandas fuctions to create and remove columns
df['Original_Bill'] = df['total_bill'] - df['tip']
df['tip percent'] = np.round((df['tip'] / df['total_bill']) * 100,2)
df.drop(columns=['smoker'], inplace=True)
print(df.head(10))
```

```python
#Statistical and Core Visualisations

#Barplot
sns.barplot(x='sex', y='size', data=df)
plt.title('Barplot')
plt.tight_layout()
plt.show()
```

```python
#Creating a boxplot
sns.boxplot(x='sex', y='total_bill', data=df)
plt.title('sex distribution by total bill')
plt.tight_layout()
plt.show()
```

```python
#ViolinPlot
sns.violinplot(x='sex', y='total_bill', data=df)
plt.title('sex distribution by total bill')
plt.show()
```

```python
#Regression Plot
sns.regplot(x='Original_Bill', y='tip', data=df)
plt.title('Original Bill vs Tip')
plt.show()
```

```python
#Heatmap of Correlation
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Map")
plt.show()
```
