```python
#CREATING PLOTS USING SEABORN AND MATPLOTLIB
#Importing
import seaborn as sns
import matplotlib.pyplot as plt

#set default style
sns.set_theme()
```

```python
#Loading the data set
df = sns.load_dataset('titanic')
df.head()
```

```python
#Creating a scatterplot
sns.scatterplot(x='age', y='fare', hue='sex', style='embark_town',data=df )
sns.set_style("whitegrid")
sns.set_palette("coolwarm")
```

```python
#Saving the plot
plt.savefig("Seaborn_plot.png", dpi=300, bbox_inches='tight')
```

```python
#Creating a histplot
sns.histplot(data=df, x='fare', kde=True)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
#Saving the histplot
plt.savefig("Seaborn Histplot",dpi=300, bbox_inches='tight')
```

```python
#Creating a barplot
plt.figure(figsize=(8,6))
sns.barplot(x='sex', y='survived', data=df, palette='pastel')
plt.title('Survival Rate by Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.tight_layout()
plt.show()
```
